```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler


# Check the transformed DataFrame
columns_to_drop = ['MinTemp', 'Pressure3pm', 'Cloud3pm', 'WindSpeed3pm', 'Evaporation', 'Sunshine', 'Temp3pm']
df = df.drop(columns=columns_to_drop)
from sklearn.impute import SimpleImputer
import pandas as pd


# Separate columns based on their data types
numerical_features = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
categorical_features = df.select_dtypes(include=['object']).columns.tolist()

# Check if categorical features exist
if categorical_features:
    # Create imputers
    categorical_imputer = SimpleImputer(strategy='most_frequent')
    df[categorical_features] = categorical_imputer.fit_transform(df[categorical_features])

# Handle numerical features only if they exist
if numerical_features:
    numerical_imputer = SimpleImputer(strategy='mean')
    df[numerical_features] = numerical_imputer.fit_transform(df[numerical_features])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OrdinalEncoder())
])

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

# Create a preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_features),
        ('num', numeric_transformer, numerical_features)
    ]
)
import pandas as pd

import pandas as pd

# Separate the Date column from df (if not already done)
df_date = df[['Date']]

# Drop the Date column from df for transformation
df_no_date = df.drop(columns=['Date'])

# Fit and transform the data (apply the transformation to the remaining columns)
df_transformed = preprocessor.fit_transform(df_no_date)

# Convert the transformed data to a DataFrame
if hasattr(preprocessor, 'get_feature_names_out'):
    # Get the transformed column names (after OneHotEncoding or similar)
    new_column_names = preprocessor.get_feature_names_out()
else:
    # For cases where no feature expansion happens, use original column names
    new_column_names = df_no_date.columns

df_transformed = pd.DataFrame(df_transformed, columns=new_column_names)

# Reattach the Date column as the first column (or wherever needed)
df_transformed.insert(0, 'Date', df_date)

# Define the new column names
desired_column_names = [
    "Date", "ID", "Location", "MaxTemp", "Rainfall",
    "WindGustDir", "WindGustSpeed", "WindDir9am", "WindDir3pm", "WindSpeed9am",
    "Humidity9am", "Humidity3pm", "Pressure9am", "Cloud9am",
    "Temp9am", "RainToday", "RainTomorrow"
]

# Ensure the length of new_column_names matches the desired_column_names
if len(df_transformed.columns) == len(desired_column_names):
    # Assign the new column names
    df_transformed.columns = desired_column_names
else:
    print(f"Mismatch in the number of columns: {len(df_transformed.columns)} in transformed data vs {len(desired_column_names)} in desired names")

# Display the resulting DataFrame
print(df_transformed.head())


# Convert the result to a DataFrame
#df_transformed = pd.DataFrame(df_transformed, columns=df_transformed.columns)
#df_transformed.head()

from sklearn.preprocessing import OrdinalEncoder
import pandas as pd


# Create an OrdinalEncoder instance
encoder = OrdinalEncoder()

# Apply the encoder to the categorical columns and assign the result back to the DataFrame
df[categorical_features] = encoder.fit_transform(df[categorical_features])

# Check the DataFrame to ensure NaN values are handled
print(df.dtypes)  # Check data types to confirm columns are handled correctly

df_transformed.head()
```

4δ.   Αξιολογήστε συνολικά την επίδοση των μοντέλων χρησιμοποιώντας κάποιο γράφημα (π.χ. ιστόγραμμα, bar plot) και σχολιάστε ποιο μοντέλο είχε την καλύτερη επίδοση.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Dictionary to store results
results = {}

# Iterate over each classifier
for clf in classifiers:
    # Make predictions on the test set
    y_pred = clf.predict(X_test)

    # Calculate performance metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # Store the results for this classifier
    results[clf.__class__.__name__] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1
    }

# Create a DataFrame from the results
df_results = pd.DataFrame(results).T.reset_index()
df_results.rename(columns={'index': 'Classifier'}, inplace=True)

# Plot the F1 Score (or any other metric you want to compare)
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.barplot(y='Classifier', x='F1 Score', data=df_results, palette='viridis')
plt.title('F1 Score Comparison Across Classifiers')
plt.xlabel('F1 Score')
plt.ylabel('Classifier')
plt.show()
```
